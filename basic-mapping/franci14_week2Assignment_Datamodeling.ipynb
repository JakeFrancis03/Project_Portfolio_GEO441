## **Reflection on data models and management in ArcGIS Pro**


  Serving as the main storage and manegement system in Geographic Informations Systems, geodatabases are made up of three seperate dataset types. These include tables, raster datasets, and feauture classes. Though each of these datasets provide different information and services to the geodatabase, feature classes act as collections (tables) of related geographic features that are based on their geometric shape in spatial data. Essentially, they are the organization of geometric shapes in spatial data. These feature classes are then grouped into feature datasets, which further organize this information. In all, geodatabases are made up of tables, raster data sets, and feature classes that provide GIS with the neccessary information and tools for spatial analysis, such as topology.
